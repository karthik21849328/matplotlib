DATA ANALYSIS -Data analysis is the process of inspecting, cleaning, transforming, and modeling data with the goal of discovering useful information, informing conclusions, and supporting decision-making.

Matplotlib is a powerful library in Python for creating visualizations. 
It provides various types of plots to suit different data visualization needs.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#LINE PLOT 
x=np.linspace(1,10,100)
y=np.exp(x)
plt.plot(x,y)
```

```python
#to represent the plot with the help of red stars 
x=np.linspace(1,10,100)
y=np.exp(x)
plt.plot(x,y,"r*")
```

```python
#python library comes with inbuilt datasets for demonstarting 
x=sns.load_dataset('tips')
x.head()
```

```python
#plotting line graph with three parameters in it and differentiated by colours 
x=np.linspace(1,10,100)
y1=x 
y2=np.square(x)
y3=np.sqrt(x)
plt.plot(x,y1,"r",x,y2,"b",x,y3,"g")
plt.title("LINE PLOT")
plt.show()
```

2. SCATTER PLOT -Scatter Plot: Used to display the relationship between two sets of data points by plotting them as individual points.

```python
tips_data=sns.load_dataset('tips')
x=tips_data['total_bill']
y=tips_data['tip']
plt.scatter(x,y)
plt.xlabel('total bill')
plt.ylabel('tip')
plt.title('Scatter plot')
plt.show()
```

```python
tips_data=sns.load_dataset('tips')
x=tips_data['total_bill']
y=tips_data['tip']
plt.scatter(x,y,s=np.array(y)**2,c="green",alpha=0.8)
plt.title("scatter with added visuals to it")
plt.show()
```

BAR PLOT-Used to represent categorical data with rectangular bars whose lengths proportional to the values they represent.

```python
tips_data.smoker.value_counts().plot(kind="bar",rot=0)
plt.title("BAR CHART OF SMOKERS")
plt.xlabel("SMOKERS")
plt.ylabel("COUNT")
```

```python
dt=sns.load_dataset('iris')
dt.head()
dt.species.value_counts().plot(kind="bar",rot=0)
plt.title("TYPE OF SPECIES")
plt.xlabel("SPECIES")
plt.ylabel("COUNT")
plt.show()
```

Pie Chart: Used to represent the proportions of different categories in a dataset as sectors of a circle.

```python
tips_data.day.value_counts()
```

```python
plt.pie(tips_data.day.value_counts(),autopct="%.1f%%",radius=1.2,labels=['sat','sun','thur','fri'],explode=[0.05,0,0,0])
plt.show()
```

Histogram: Used to represent the distribution of a continuous variable by dividing the data into intervals (bins) and displaying the frequency of observations in each bin.
* Histogram work only on numerical data
* Bar plot will work on categorical data
* pie plot will work on propetion or percentages 

```python
plt.hist(tips_data['tip'],bins=10)
```

Box Plot: Used to visualize the distribution of a dataset by displaying the five-number summary (minimum, first quartile, median, third quartile, and maximum) and any outliers.m

```python
plt.boxplot(tips_data['total_bill'])
plt.text(x=1.1,y=tips_data["total_bill"].min(),s="min")
plt.text(x=1.1,y=tips_data["total_bill"].max(),s="max")
plt.text(x=1.1,y=tips_data["total_bill"].median(),s="median")
plt.text(x=1.1,y=tips_data.total_bill.quantile(0.75),s="q1")
plt.text(x=1.1,y=tips_data.total_bill.quantile(0.25),s="q1")
```
